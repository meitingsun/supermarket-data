### 1、销售额

#### 1.1 每年各省的销售额

```python
data0=data # 复制一份原始数据 
data0['年']=data0['订单日期'].dt.year
data0['月']=data0['订单日期'].dt.month
print('数据类型：',data0.dtypes)
data0.head()
```

```python
sales=data.groupby(['省/自治区','年'])['销售额'].sum()
sales=sales.unstack()
from pyecharts import Geo,Timeline 

sales_2015=sales[2015]
sales_2016=sales[2016]
sales_2017=sales[2017]
sales_2018=sales[2018]

# 2015年各省销售额分布情况
attr_2015=sales_2015.index
value_2015=sales_2015.values
geo_2015=Geo('2015年各省销售总额',title_color='#0A0A0A',width=700, height=400, background_color='#FFFFFF')
geo_2015.add('销售额 ',attr_2015,value_2015,visual_range=[min(value_2015),max(value_2015)], visual_text_color="#0A0A0A", symbol_size=10,is_visualmap=True)

# 2016年各省销售额分布情况
attr_2016=sales_2016.index
value_2016=sales_2016.values
geo_2016=Geo('2016年各省销售总额',title_color='#0A0A0A',width=700, height=400, background_color='#FFFFFF')
geo_2016.add('销售额 ',attr_2016,value_2016,visual_range=[min(value_2016),max(value_2016)])

# 2017年各省销售额分布情况
attr_2017=sales_2017.index
value_2017=sales_2017.values
geo_2017=Geo('2017年各省销售总额',title_color='#0A0A0A',width=700, height=400, background_color='#FFFFFF')
geo_2017.add('销售额 ',attr_2017,value_2017,visual_range=[min(value_2017),max(value_2017)])
# 2018年各省销售额分布情况
attr_2018=sales_2018.index
value_2018=sales_2018.values
geo_2018=Geo('2018年各省销售总额',title_color='#0A0A0A',width=700, height=400, background_color='#FFFFFF')
geo_2018.add('销售额 ',attr_2018,value_2018,visual_range=[min(value_2018),max(value_2018)])

# 添加时间线
timeline=Timeline(is_auto_play=False,timeline_bottom=0)
timeline.add(geo_2015,'2015年')
timeline.add(geo_2016,'2016年')
timeline.add(geo_2017,'2017年')
timeline.add(geo_2018,'2018年')
timeline.render(path=r"C:\Users\23691\AnacondaProjects\Supermarket Data\销售额\每年各省.html")
# timeline.render(r'每年各省.html')
timeline
#geo  # 在notebook上直接显示
```

#### title_color是标题的颜色；background_color是背景颜色；visual_text_color

#### 1.2  每年各类别销售额分布

```python
sales_type=data0.groupby(['年','类别'])['销售额'].sum().unstack()
from pyecharts import Bar
Bar=Bar('每年各类别销售额分布图','总销售额在逐年增加')
Bar.add('办公用品',sales_type.index,sales_type['办公用品'].values,is_stack=True)
Bar.add('家具',sales_type.index,sales_type['家具'].values,is_stack=True)
Bar.add('技术',sales_type.index,sales_type['技术'].values,is_stack=True)
Bar.render(path=r"C:\Users\23691\AnacondaProjects\Supermarket Data\销售额\每年各类别.html")
Bar
```

#### 2.2 每年各客户总的订购数量

```python
num_order=data0.groupby(['年','客户名称'])['数量'].sum().unstack()
num_order=num_order.T

# 填充缺失值 补充为0
num_order=num_order.fillna(0)

# 多列进行 加和运行
num_order['总销售量']=num_order.apply(lambda x:x[2015]+x[2016]+x[2017]+x[2018],axis=1)


num_order['订购增长率']=num_order.apply(lambda x:(x[2018]-x[2015])/x[2015],axis=1)

# 总订购量 top25
num_top=pd.DataFrame(num_order['总销售量']).sort_values(by=['总销售量'],ascending=False).head(25)

rate_top=pd.DataFrame(num_order['订购增长率']).sort_values(by=['订购增长率'],ascending=False).head(25)
```

#### 3、发货时长

```python
data0['发货时长']=data0.apply(lambda x:x['发货日期']-x['订单日期'],axis=1)
print("数据类型",data0.dtypes)  # 时间段（Timedelta），可以算 时，分等时间段
print("*****************************")
data0.head()
```

```python
# 提取两个datatime差-->时间段的数值
data0['发货时长']=data0['发货时长'].dt.days
data0.head(1)
```

```python
# 每年各子类总的发货时长
sum_time=data0.groupby(['年','子类别'])['发货时长'].sum().unstack()

#每年各子类平均发货时长
mean_time=data0.groupby(['年','子类别'])['发货时长'].mean().unstack()
```
